### Linear Regression

Linear Equation 

y = b0 + b1*x where b0 and b1 reperent the intercept and slope. 

Measuring the coefficient. Least Square approach.
RSS - Residual sun of square. 
Residual is the difference between y - y (observed vs predicted)





Assessing the accuracy of the coefficient estimate. 
True relationship between X and Y =: y = f(X) + e where e is mean zero random error term. 


```python
# Import libraries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
rng = np.random.default_rng()

x = rng.random(100)
# x = rng.normal(size = (100))
# y = rng.normal(size = (100))
e = rng.normal(size = (100))
```

```python
x
```

```python
y = 2 + 3*x + e
y
```

```python
plt.hist(y)
```

```python
plt.plot(x)
```

```python
plt.plot(y)
```

```python
plt.plot(x,y,)
```

```python
plt.scatter(x,y)
```

```python
# np.polynomial fit

m_slope, b_intercept = np.polynomial.polynomial.polyfit(x, y , 1)
m_slope, b_intercept
```

```python
plt.plot(x,y, 'o', color='green')
plt.plot(x, 2 + 3*x, color='red') # True relationship known as population regression line.
plt.plot(x, m_slope*x + b_intercept, color='blue') # Least square estimate for f(X)
```

```python
# Visualizing using seaborn library
sns.jointplot(x=x, y=y, ci=None, kind='reg')
```

### Calculate Residual Standard Error; R Square; F - Statistics

```python
# Residual = Observed value and predicted value.

# using scikit learn library 

from sklearn.linear_model import LinearRegression

model = LinearRegression()
x = ad_data[ad_data.columns[0:3]]
y = ad_data.Sales
model.fit(x,y)

y_pred = model.predict(x)

df = pd.DataFrame({'Actual': y, 
                  "Predicted": y_pred})

# print(df)

print(f"Residual sum of Square " + str(np.sum(np.square(df['Actual'] - df['Predicted']))))
```

```python
# using statmodel.api
import statsmodels.api as sm

x = sm.add_constant(x)
model = sm.OLS(y, x).fit()

print(model.summary())
print(model.ssr)
```
